```python
import numpy as np

# Load the .npz file
data = np.load("D:/BR/CGAN-dpi/cloudsat-gan-master/cloudsat-gan-master/winduvcgan_v3/data/scenes_output.npz")

# Check the keys in the loaded file
print(data.files)
```

```python
# Access the data arrays
scene_initial = data['scene_initial']
scene_gen = data['scene_gen']
scene_real = data['scene_real']
scene_dispre = data['scene_dispre']
```

```python
print(scene_initial.shape)
print(scene_gen.shape)
print(scene_real.shape)
print(scene_dispre.shape)
```

```python
rmse_per_channel = []
for c in range(scene_real.shape[-1]):  # 遍历两个通道
    diff = scene_gen[..., c] - scene_real[..., c]
    non_zero_mask = diff != 0
    squared_diff = (diff ** 2) * non_zero_mask
    mse = np.sum(squared_diff) / np.sum(non_zero_mask)
    rmse_per_channel.append(np.sqrt(mse))

print("每个通道的RMSE:", rmse_per_channel)
```

```python
squared_diff = (scene_gen - scene_real) ** 2
mse = np.mean(squared_diff, axis=(0, 1, 2))
# 计算RMSE，即均方误差的平方根
rmse = np.sqrt(mse)
```
